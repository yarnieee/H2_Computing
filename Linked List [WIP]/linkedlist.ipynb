#### Node for Linked List

```python
# node for linked list

class Node:
	def __init__(self, data, next=None):
		self.data=data
		self.next=next
```

#### Linked list (iterative)
##### possible objectives (not exhaustive)
1. insert at a position
2. insert in a certain order (ie. alphabetical)
3. delete at a position
4. delete first instance of a value
5. delete all instances of a value
6. delete all which fulfil a criteria (i.e. value < 5)
7. push/pop functions of a stack/queue

#### Binary Search Tree

```python
class Node:
	def __init__(self,data,left=None,right=None):
		self.data=data
		self.left=left
		self.right=right
		
	def __str__(self):
		if self.left is not None:
			left=self.left.data
		else:
			left=None

		if self.right is not None:
			right=self.right.data
		else:
			right=None

		return f"Data: {self.data}, Left: {left}, Right:  {right}"
	
class BST():
	def __init__(self):
		self.root=None
		
	# add method
	def add(self, key): 
		self.root=self._add(self.root, key)
	
	def _add(self,currnode,key):
		if currnode is None:
			return Node(key)
		if key > currnode.data:
			currnode.right=self._add(currnode.right, key)
		elif key < currnode.data:
			currnode.left=self._add(currnode.left, key)
		return currnode
	
	#add WITHOUT RECURSION
	def insert(self, value):
		if self.root is None:
			self.root=Node(value)
			return
		else:
			curr_node=self.root
		
		#find the node where the new value is supposed to go
		while curr_node is not None:
			if value > curr_node.data: #value greater than current node
				if curr_node.right is None:
					curr_node.right=Node(value)
				else:
					curr_node=curr_node.right
			elif value < curr_node.data: #value smaller than current node
				if curr_node.left is None:
					curr_node.left=Node(value)
				else:
					curr_node=curr_node.left
			else:
				return
		#changes value for current node
	
	
	# in order traversal
	def _inorder(self, currnode):
		#helper
		if currnode is not None:
			self._inorder(currnode.left)
			print(currnode.data, end=" ")
			self._inorder(currnode.right)

	def inorder(self):
		print("In order:",end=" ")
		self._inorder(self.root)
		
		
	# preorder traversal
	def preorder(self):
		print("preorder: ",end="")
		self._preorder(self.root)
	
	def _preorder(self, currnode): #root left right
		if currnode is not None:
			print(currnode.data,end=" ")
			self._preorder(currnode.left)
			self._preorder(currnode.right)
			
			
	# preorder traversal (stores values in list instead of direct print)
	def preorder2(self):
		result=[]
		self._preorder2(self.root, result)
		return result
		
	def _preorder2(self, currnode, result):
		if currnode is not None:
			result.append(currnode.data)
			self._preorder2(currnode.left, result)
			self._preorder2(currnode.right, result)
			
			
	# postorder
	def postorder(self):
		self._postorder(self.root)
	
	def _postorder(self, currnode):
		if currnode is not None:
			self._postorder(currnode.left)
			self._postorder(currnode.right)
			print(currnode.data,end=" ")
	
	
	# find
	def find(self, value):
		return self._find(self.root, value)
			
	def _find(self, curr_node, value):
		if value==curr_node.data: #compare to root
			return curr_node #root is a node
		elif value < curr_node.data: #recurse _find left
			if curr_node.left is not None:
				return self._find(curr_node.left, value)
			else:
				return None
		else: #recurse _find right
			if curr_node.right is not None: #is this the only method or
				return self._find(curr_node.right, value)
			else:
				return None
			
	
	# search
	def search(self, value):
		if self.root is None:
			return None
		else:
			curr_node=self.root
		
		#find the node where the new value is supposed to go
		while curr_node is not None:
			if value > curr_node.data:
				if curr_node.right is not None:
					curr_node=curr_node.right
				else:
					return None
			elif value < curr_node.data:
				if curr_node.left is not None:
					curr_node=curr_node.left
				else:
					return None
			else:
				return curr_node #value matches current node
```
